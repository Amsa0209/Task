```python
n=len(data_test)
n
```

```python
start_index= n-12
start_index
```

```python
df=data_test.iloc[:start_index]
df_t=data_test.iloc[start_index:]
df.tail()
```

```python
len(df)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model=SARIMAX(df["Sales"],order=(2,0,0),seasonal_order=(1,1,0,12))
results=model.fit()
results.summary()
```

```python
start=len(df)
end=len(df)+len(df_t)-1
predictions= results.predict(start=start, end=end+12, dynamic=False, type="level")
```

```python
predictions
```

```python
predictions.plot(x="Sales",y="predicted_mean")
plt.show()
```
