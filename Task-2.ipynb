```python
n=len(df)
n
```

```python
start_index= n-12
start_index
```

```python
df1=df.iloc[:start_index]
df1_t=df.iloc[start_index:]
df1.tail()
```

```python
len(df)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model=SARIMAX(df["Rating"],order=(2,0,0),seasonal_order=(1,1,0,12))
results=model.fit()
results.summary()
```

```python
start=len(df1)
end=len(df1)+len(df1_t)-1
predictions= results.predict(start=start, end=end+12, dynamic=False, type="level")
```

```python
predictions
```

```python
predictions.plot(x="Rating",y="predicted_mean")
plt.show()
```
